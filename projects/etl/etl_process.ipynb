# Data Loading

Importing packages

```python
import pandas as pd
import numpy as np
```

Loading data from Excel

```python
# file_path = "data/assessment_dataset.xlsx"
file_path = "data/assessment_dataset.csv"

sales = pd.read_csv(file_path)

# sales.to_csv('data/assessment_dataset.csv', index=False)
```

Understanding Data

```python
sales.shape
```

```python
sales.head(n=10)
```

```python
sales.dtypes
```

Transforming Returned column data type to boolean

```python
sales['Returned'].unique()
```

```python
return_mapping = {'Yes': True, 'No': False}
sales['Returned'] = sales['Returned'].map(return_mapping)

sales['Returned']
```

```python
sales['IsPromotional'].unique()
```

Transforming IsPromotional column data type to boolean

```python
return_mapping = {'Yes': True, 'No': False}
sales['IsPromotional'] = sales['IsPromotional'].map(return_mapping)

sales['IsPromotional']
```

```python
sales['TransactionDate'] = pd.to_datetime(sales['TransactionDate'])
sales['TransactionDate']
```

```python
# Set the 'TransactionID' column as the index
sales.set_index('TransactionID', inplace=True)
```

Checking Null Values

```python
sales.isna().sum()
```

```python
sales.isna().sum(axis=1)
```

```python
sales.isna().sum().sum()
```

Forward Filling TransactionDate column


```python
# TODO: ffill for date here, because dates are in sorted and very short range, may be usefull for ML training over just droping it.

sales['TransactionDate'] = sales['TransactionDate'].ffill()
sales['TransactionDate'].isna().sum()
```

Dropping all rows with null

```python
sales = sales.dropna()

sales.shape
```

```python
sales.isna().sum()
```

Data Quality Checking

```python
sales.duplicated().sum()
```

```python
sales.index.duplicated().sum()
```

```python
sales[(sales['CustomerAge'] < 0) | (sales['CustomerAge'] > 100)]
```

```python
sales[sales['TransactionAmount'] < 0]
```

```python
sales = sales[sales['TransactionAmount'] >= 0]
```

```python
sales[(sales['DiscountPercent'] < 0) | (sales['DiscountPercent'] > 100)]
```

```python
sales[sales['Quantity'] <= 0]
```

Loading: Writing Clean Data to CSV 

```python
sales.to_csv("data/clean_assessment_dataset.csv")
```

Loading: in SQL Server

```python
import pandas as pd
from sqlalchemy import create_engine
import urllib
```

```python
SERVER = "localhost"
DATABASE = "etl_practice_db"
DRIVER = "ODBC Driver 17 for SQL Server"

params = urllib.parse.quote_plus(
    f"DRIVER={{{DRIVER}}};SERVER={SERVER};DATABASE={DATABASE};Trusted_Connection=yes;TrustServerCertificate=yes"
)
engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")
```

```python
sales.to_sql("clean_sales_data", con=engine, if_exists="replace")
```
